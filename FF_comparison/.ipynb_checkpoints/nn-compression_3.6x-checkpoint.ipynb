# Loading dependencies

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from tqdm import trange
```

# Loading cube

10, 1536, 1536, 35 = shape of the data

10 - physical parameters (T, p, B(x,y,z), v(x,y,z), tau, z_axis)\
1536 x 1536 - field \
35 - z layers 

```python
fits_path = '/kaggle/input/datasets/siladavid/muram-iso-z-second-batch/muram_3D_cutout1800.fits'
data_raw = fits.open(fits_path)

print("Broj HDU:", len(data_raw))
print("Shape:", data_raw[0].data.shape)
```

```python
SUB = 64 # taking the sub domen of the cube 1536x1536 -> 128x128

T = data_raw[0].data[:SUB, :SUB, :]
p = data_raw[1].data[:SUB, :SUB, :]
Bx = data_raw[2].data[:SUB, :SUB, :]
By = data_raw[3].data[:SUB, :SUB, :]
Bz = data_raw[4].data[:SUB, :SUB, :]
vx = data_raw[5].data[:SUB, :SUB, :]
vy = data_raw[6].data[:SUB, :SUB, :]
vz = data_raw[7].data[:SUB, :SUB, :]
tau = data_raw[8].data[:SUB, :SUB, :]

z_axis = data_raw[9].data
```

```python
print("T shape:", T.shape) # (Nx, Ny, Nz)
print("z_axis shape:", z_axis.shape)
print("z_axis range:", z_axis.min(), z_axis.max())

fields = [T, p, Bx, By, Bz, vx, vy, vz]

field_names = ['T', 'p', 'Bx', 'By', 'Bz', 'vx', 'vy', 'vz']

for name, f in zip(field_names, fields):
    print(f"{name}: shape={f.shape}, min={f.min():.3e}, max={f.max():.3e}")
```

# Normalizing x,y,z

```python
Nx, Ny, Nz = T.shape

x_coords = np.linspace(0, 1, Nx)
y_coords = np.linspace(0, 1, Ny)
z_coords = np.linspace(0, 1, Nz)

xx, yy, zz = np.meshgrid(x_coords, y_coords, z_coords, indexing='ij')

coords = np.stack([xx.ravel(), yy.ravel(), zz.ravel()], axis=1).astype(np.float32)
print("coords shape:", coords.shape)
```

# Normalizing targets

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device we are using:", device)

targets_raw = np.stack([f.ravel() for f in fields], axis=1).astype(np.float32)  # (N, 8)
print('Targets shape:', targets_raw.shape)

# z-score
field_mu  = targets_raw.mean(axis=0)
field_sd  = targets_raw.std(axis=0) + 1e-8
targets_norm = (targets_raw - field_mu) / field_sd
```

## Mixed normalization for B and p

```python
X_train = torch.from_numpy(coords).to(device)
Y_train = torch.from_numpy(targets_norm).to(device)

print("Training on:", X_train.shape[0])
```

```python
print("z_axis:", z_axis.shape, z_axis.min(), z_axis.max(), 
      "monotono rastuce?" , bool(np.all(np.diff(z_axis) > 0)))
for name, f in zip(['T','p','Bx','By','Bz','vx','vy','vz'], fields):
    print(f"{name:3s} min={f.min():.3e}  max={f.max():.3e}  median={np.median(f):.3e}")
print("tau:", tau.shape, tau.min(), tau.max())  # imas i tau iz HDU 8
```

# Defining neural network structure

```python
'''
class NeuralField(nn.Module):
    
    def __init__(self, hidden_sizes=[256, 256, 256, 256, 256]):
        super().__init__()
        layers = []
        sizes = [3] + hidden_sizes + [8]
        for i in range(len(sizes) - 1):
            
            layers.append(nn.Linear(sizes[i], sizes[i+1]))
            
            if i < len(sizes) - 2:
                layers.append(nn.Tanh())
                
        self.net = nn.Sequential(*layers)
        
        for m in self.net:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        return self.net(x)

'''
```

```python
###
class FourierFeatures(nn.Module):
    def __init__(self, freqs_per_axis=(7, 7, 5), scale=2.0):
        super().__init__()
        
        rows = [(ax, scale**k * np.pi)
                for ax, F in enumerate(freqs_per_axis) for k in range(F)]
        
        self.register_buffer("ax_idx", torch.tensor([r[0] for r in rows]))
        self.register_buffer("freq", torch.tensor([r[1] for r in rows], dtype=torch.float32))
        self.out_dim = 3 + 2 * len(rows)        # raw + sin + cos

    def forward(self, x):                        # x: (..., 3)
        proj = x[..., self.ax_idx] * self.freq
        return torch.cat([x, torch.sin(proj), torch.cos(proj)], dim=-1)

###
class NeuralField(nn.Module):
    def __init__(self, width=344, depth=8, freqs_per_axis=(7, 7, 5)):
        
        super().__init__()
        self.enc = FourierFeatures(freqs_per_axis)
        d = self.enc.out_dim
        self.skip_at = depth // 2

        self.layers = nn.ModuleList()
        self.layers.append(nn.Linear(d, width))
        
        for i in range(1, depth):
            in_w = width + d if i == self.skip_at else width  # skip
            self.layers.append(nn.Linear(in_w, width))
            
        self.out = nn.Linear(width, 8)
        self.act = nn.Tanh()

        for m in self.modules():
            if isinstance(m, nn.Linear):
                
                nn.init.xavier_uniform_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        h0 = self.enc(x)
        h = h0
        for i, lin in enumerate(self.layers):
            if i == self.skip_at:
                h = torch.cat([h, h0], dim=-1)
            h = self.act(lin(h))
        return self.out(h)
```

```python
import gc
torch.cuda.empty_cache()
gc.collect()

model = NeuralField(width=210, depth=8, freqs_per_axis=(7,7,5)).to(device)
```

```python
n_params = sum(p.numel() for p in model.parameters())
n_target = targets_raw.size

print(f"No. of network parameters: {n_params:,}")
print(f"Target size: {n_target:,}")
print(f"Param ratio: {n_target / n_params:.1f}x")
```

```python
batch_size        = 200_000
n_train           = X_train.shape[0]
BATCHES_PER_EPOCH = 10   
history_loss      = []
epochs = 3000
best = np.inf; patience = 0; PATIENCE_LIMIT = 300

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
```

# Training loop

```python
for epoch in trange(epochs):
    epoch_loss = 0.0
    for _ in range(BATCHES_PER_EPOCH):
        idx     = torch.randint(0, n_train, (batch_size,), device=device)
        X_batch = X_train[idx]
        Y_batch = Y_train[idx]

        optimizer.zero_grad()
        loss = nn.functional.mse_loss(model(X_batch), Y_batch)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    scheduler.step()
    avg_loss = epoch_loss / BATCHES_PER_EPOCH
    history_loss.append(avg_loss)

    if epoch % 200 == 0:
        print(f"Epoch {epoch} | Loss: {avg_loss:.4e} | LR: {scheduler.get_last_lr()[0]:.2e}")
        
    if avg_loss < best - 1e-6:
        best = avg_loss; patience = 0
        torch.save({'model_state': model.state_dict(),
                    'field_mu': field_mu, 'field_sd': field_sd,
                    'arch': {...}}, '/kaggle/working/best.pt')   # cuva najbolji
    else:
        patience += 1
        if patience >= PATIENCE_LIMIT:
            print(f"Early stop na epohi {epoch}"); break
```

# Calling the model

```python
# Full grid coordinates
X_full = torch.from_numpy(coords).to(device)

model.eval()
chunk = 200_000
preds = []

with torch.no_grad():
    for s in range(0, X_full.shape[0], chunk):
        preds.append(model(X_full[s:s+chunk]).cpu().numpy())

pred_norm = np.concatenate(preds, axis=0)
print("pred_norm shape:", pred_norm.shape)

# Denormalization
pred_raw = pred_norm * field_sd + field_mu
pred_raw = pred_raw.reshape(Nx, Ny, Nz, 8)
true_raw = targets_raw.reshape(Nx, Ny, Nz, 8)
```

# Ploting the results

```python
iz = Nz // 2  # middle z layer

fig, axes = plt.subplots(8, 3, figsize=(15, 40))

for i, name in enumerate(field_names):
    
    # 2D maps on middle z
    
    f_true = true_raw[:, :, iz, i]   # (Nx, Ny)
    f_pred = pred_raw[:, :, iz, i]   # (Nx, Ny)

    vmin = f_true.min()
    vmax = f_true.max()

    # --- Plot 1: Truth ---
    ax = axes[i, 0]
    im = ax.imshow(f_true, origin='lower', vmin=vmin, vmax=vmax, cmap='plasma')
    ax.set_title(f'{name} — Original')
    ax.set_xlabel('x'); ax.set_ylabel('y')
    plt.colorbar(im, ax=ax)

    # --- Plot 2: Pred ---
    ax = axes[i, 1]
    im = ax.imshow(f_pred, origin='lower', vmin=vmin, vmax=vmax, cmap='plasma')
    ax.set_title(f'{name} — Compressed')
    ax.set_xlabel('x'); ax.set_ylabel('y')
    plt.colorbar(im, ax=ax)

    # --- Plot 3: Scatter 1:1 ---
    ax = axes[i, 2]
    
    # Subsample for scatter plot
    n_samples = min(5000, f_true.size)
    idx = np.random.choice(f_true.size, size=n_samples, replace=False)
    ax.scatter(f_true.ravel()[idx], f_pred.ravel()[idx],
               alpha=0.3, s=5, color='steelblue')
    # Linija 1:1
    ax.plot([vmin, vmax], [vmin, vmax], 'r-', linewidth=1.5, label='1:1')
    ax.set_xlabel(f'{name} original')
    ax.set_ylabel(f'{name} compressed')
    ax.set_title(f'{name} — Scatter')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

fig.suptitle(f'2D maps and scatter — z layer {iz} from {Nz}', fontsize=16, y=1.001)
plt.tight_layout()
plt.show()
```

```python
for i, name in enumerate(field_names):
    t = true_raw[..., i]; p = pred_raw[..., i]
    rmse = np.sqrt(np.mean((p - t)**2))
    rel_std   = rmse / t.std() * 100                    
    rel_range = rmse / (t.max() - t.min()) * 100        
    print(f"{name}: RMSE={rmse:.3e}  rel_std={rel_std:5.2f}%  rel_range={rel_range:4.2f}%")
```

```python
iT, ivz, iBz = 0, 7, 4

def tau_weight(tau, tau_center=1.0, sigma=0.5):
    lt = np.log10(tau + 1e-30)
    w = np.exp(-0.5*((lt - np.log10(tau_center))/sigma)**2)
    return w / (w.sum(axis=2, keepdims=True) + 1e-30)

def moment(cube, ch, w):
    return (cube[..., ch] * w).sum(axis=2)   # (Nx,Ny)


for tau_c in [1.0, 0.1]:
    w = tau_weight(tau, tau_center=tau_c)
    print(f"\n=== tau_center = {tau_c} (proxy for {'continumm' if tau_c==1 else 'line core'}) ===")
    for ch, name, what in [(iT,'T','continumm'),
                           (ivz,'vz','Doppler shift of core'),
                           (iBz,'Bz','Stokes V')]:
        mt = moment(true_raw, ch, w)
        mp = moment(pred_raw, ch, w)
        rel = np.sqrt(np.mean((mp-mt)**2)) / (mt.std()+1e-30) * 100
        bias = (mp.mean()-mt.mean())/(mt.std()+1e-30)*100   
        print(f"  {name:3s} ({what:22s}): rel_std={rel:5.2f}%  bias={bias:+5.2f}%")
```

```python
w = tau_weight(tau, tau_center=1.0)
mt = moment(true_raw, ivz, w); mp = moment(pred_raw, ivz, w)

fig, ax = plt.subplots(1, 3, figsize=(15, 4))
vlim = np.abs(mt).max()
ax[0].imshow(mt, origin='lower', cmap='RdBu_r', vmin=-vlim, vmax=vlim); ax[0].set_title('vz τ-weighted — truth')
ax[1].imshow(mp, origin='lower', cmap='RdBu_r', vmin=-vlim, vmax=vlim); ax[1].set_title('vz τ-weighted — pred')
im = ax[2].imshow(mp-mt, origin='lower', cmap='RdBu_r'); ax[2].set_title('error'); plt.colorbar(im, ax=ax[2])
plt.tight_layout(); plt.show()
```

```python
def radial_power(img):
    F = np.abs(np.fft.fftshift(np.fft.fft2(img)))**2
    cy, cx = np.array(F.shape)//2
    y, x = np.indices(F.shape)
    r = np.hypot(x-cx, y-cy).astype(int)
    tbin = np.bincount(r.ravel(), F.ravel())
    nbin = np.bincount(r.ravel())
    return tbin / np.maximum(nbin, 1)

iz = Nz // 2
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for i, name in enumerate(field_names):
    pt = radial_power(true_raw[:, :, iz, i])
    pp = radial_power(pred_raw[:, :, iz, i])
    k = np.arange(len(pt))
    ax = axes.flat[i]
    ax.loglog(k[1:], pt[1:], 'k-', label='truth', lw=1.5)
    ax.loglog(k[1:], pp[1:], 'r--', label='pred', lw=1.5)
    ax.set_title(name); ax.set_xlabel('k (radial)'); ax.grid(alpha=0.3, which='both')
    if i == 0: ax.legend()
        
fig.suptitle(f'Spatial power spectrum, z={iz}')
plt.tight_layout(); plt.show()
```

```python
print("=== Vertical gradient (dQ/dz) — rel_std ===")

for i, name in enumerate(field_names):
    gt = np.gradient(true_raw[..., i], axis=2)
    gp = np.gradient(pred_raw[..., i], axis=2)
    rel = np.sqrt(np.mean((gp-gt)**2)) / (gt.std()+1e-30) * 100
    print(f"  d{name}/dz: rel_std = {rel:5.2f}%")
```

```python
print("=== Error for |Bz| ===")
absBz = np.abs(true_raw[..., 4]).ravel()
err   = (pred_raw[..., 4] - true_raw[..., 4]).ravel()
for lo, hi in [(0,50),(50,200),(200,500),(500,2000)]:
    m = (absBz>=lo)&(absBz<hi)
    if m.sum()>0:
        rel = np.sqrt(np.mean(err[m]**2)) / (true_raw[...,4].std()) * 100
        print(f"  |Bz| {lo:4d}-{hi:4d} G  ({m.sum():6d} tac): rel_std={rel:5.2f}%")
```
